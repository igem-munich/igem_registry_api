```python
import os

from dotenv import load_dotenv

from igem_registry_api import Client

client = Client()
client.connect()

load_dotenv()
username = os.getenv("IGEM_USERNAME")
password = os.getenv("IGEM_PASSWORD")

if username and password:
    client.sign_in(
        username=username,
        password=password,
    )

if not client.is_opted_in:
    client.opt_in()
```

```python
from rich.pretty import pprint

from igem_registry_api import Account

user = client.me()
pprint(user)

account = Account(
    client=client,
    uuid=user.uuid,
    first_name=user.first_name,
    last_name=user.last_name,
    photo=user.photo,
    consent=user.consent,
)
```

```python
orgs = user.affiliations(limit=2)
pprint(orgs)
```

```python
parts = user.parts(limit=10)
pprint(parts)
```

```python
for org in orgs:
    members = org.members()
    pprint(members)
```
